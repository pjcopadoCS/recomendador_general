```python
import pandas as pd
from googletrans import Translator

# Create a sample DataFrame with a column of words to translate
df = pd.read_excel('wines_dataset.xlsx')

# Initialize the translator
translator = Translator()

# Function to translate words to English
def translate_to_english(word):
    return translator.translate(word, src='auto', dest='en').text

# Apply the translation to the DataFrame column
for col in df.columns:
    df[col+"_en"] = df[col].apply(translate_to_english)

# Show the DataFrame with the translations
print(df)
```
